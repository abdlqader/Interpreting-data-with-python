```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

automobile_df = pd.read_csv('archive/Automobile_data.csv',na_values='?')
automobile_df.head()
```

```python
automobile_df.isnull().sum()
```

```python
automobile_df.dropna(inplace=True)
automobile_df.shape
```

```python
automobile_df.to_csv('archive/automobile_processed.csv')
```

```python
plt.figure(figsize=(12,8))

sns.distplot(automobile_df['price'],color='red')

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.distplot(automobile_df['price'],color='red',hist=False,rug=True)

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.rugplot(automobile_df['price'],height=0.5,color='Blue')

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.kdeplot(automobile_df['price'],shade=True,color='red')

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(automobile_df['engine-size'],automobile_df['price'],hue=automobile_df['num-of-cylinders'],s=120)

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.regplot(data=automobile_df,x='horsepower',y='price')

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.regplot(data=automobile_df,x='highway-mpg',y='price')

plt.title('Automobile Data')
plt.show()
```

```python
sns.jointplot(data=automobile_df,x='horsepower',y='price')
plt.show()
```

```python
sns.jointplot(data=automobile_df,x='horsepower',y='price',kind='reg')
plt.show()
```

```python
sns.jointplot(data=automobile_df,x='horsepower',y='price',kind='kde')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(6,6))

sns.kdeplot(automobile_df['horsepower'],automobile_df['price'],ax=ax)

sns.rugplot(automobile_df['horsepower'],color='limegreen',ax=ax)
sns.rugplot(automobile_df['price'],color='red',vertical=True,ax=ax)

plt.title('Automobile Data')

plt.show()
```
