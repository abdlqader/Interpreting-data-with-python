```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
bikesharing_df = pd.read_csv('archive/bike_sharing_hourly.csv')
bikesharing_df.head()
```

```python
bikesharing_df['temp'].autocorr(lag=2)
```

```python
bikesharing_df['temp'].autocorr(lag=24)
```

```python
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(12,14))

ax1.acorr(bikesharing_df['temp'],maxlags=12, color='green')
ax1.title.set_text('Temperature')
ax1.set_xlabel('Lags',fontsize=15)

ax2.acorr(bikesharing_df['hum'],maxlags=12,color='red')
ax2.title.set_text('Humidity')
ax2.set_xlabel('lags',fontsize=15)

plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(12,14))

ax1.acorr(bikesharing_df['temp'],maxlags=48,color='g')
ax1.title.set_text('Temperature')
ax1.set_xlabel('Lags',fontsize=15)

ax2.acorr(bikesharing_df['hum'],maxlags=48,color='r')
ax2.title.set_text('Humidity')
ax2.set_xlabel('Lags',fontsize=15)
```
