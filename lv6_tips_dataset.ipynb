```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

tips_df = pd.read_csv('archive/tips.csv')
tips_df.head()
```

```python
tips_df.describe()
```

```python
tips_df.groupby('time').mean()
```

```python
sns.barplot(data=tips_df,x='time',y='total_bill')
```

```python
mean = tips_df['tip'].mean()
tips_df['above average'] =  (tips_df['tip']-mean )>0
tips_df[['tip','above average']].head()
```

```python
plt.figure(figsize=(12,8))

sns.countplot('time',hue='above average',data=tips_df,order = tips_df['time'].value_counts().index)
```

```python
plt.figure(figsize=(12,8))

sns.countplot(data=tips_df,x='time',hue='sex')
```

```python
plt.figure(figsize=(12,8))

sns.boxplot(data=tips_df,x='size',y='total_bill',)
```

```python
sns.catplot(data=tips_df,x='size',y='total_bill',hue='time',kind='bar',height=6,aspect=2,palette='CMRmap')
plt.title('Bill and Tips')
```

```python
plt.figure(figsize=(12,8))

sns.stripplot(x='size',y='tip',data=tips_df)
plt.title('Bill and tips')
```

```python
 plt.figure(figsize=(12,8))

sns.swarmplot(x='size',y='tip',data=tips_df)
plt.title('Bill and tips')
```

```python
plt.figure(figsize=(12,8))

sns.swarmplot(x='size',y='total_bill',hue='above average',data=tips_df)
plt.title('Bill and tips')
```

```python
plt.figure(figsize=(12,8))

sns.swarmplot(x='size',y='total_bill',hue='above average',dodge=True,data=tips_df)
plt.title('Bill and tips')
```
